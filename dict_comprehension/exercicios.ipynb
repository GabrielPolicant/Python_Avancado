Exercício 1 - Quadrado dos números

Dada a lista:

numeros = [1, 2, 3, 4, 5]

Crie um dicionário onde:

a chave seja o número;
o valor seja o quadrado desse número.

```python
numeros = range(6)
quadrados = {n: n**2 for n in numeros}
print(quadrados)
```

Exercício 2 - Apenas números pares

Dada a lista:

numeros = list(range(1, 11))

Utilize Dict Comprehension para criar um dicionário contendo apenas os números pares.

```python
numeros = list(range(1, 11))
pares = {n: n for n in numeros if n % 2 == 0}
print(pares)
```

Exercício 3 - Aumentando preços

Dado o dicionário:

produtos = {
    "Notebook": 3500,
    "Mouse": 80,
    "Teclado": 200,
    "Monitor": 900
}

Crie um novo dicionário onde todos os preços recebam um aumento de 15%.

```python
produtos = {
    "Notebook": 3500,
    "Mouse": 80,
    "Teclado": 200,
    "Monitor": 900
}

novos_precos = {produto: round(preco * 1.15, 2) for produto, preco in produtos.items()}
print(novos_precos)
```

Exercício 4 - Classificando notas

Dado o dicionário:

alunos = {
    "Gabriel": 9,
    "Lucas": 5,
    "Ana": 8,
    "Marcos": 4,
    "Julia": 7
}

Crie um novo dicionário onde:

notas maiores ou iguais a 7 → "Aprovado"
notas menores que 7 → "Reprovado"

```python
alunos = {
    "Gabriel": 9,
    "Lucas": 5,
    "Ana": 8,
    "Marcos": 4,
    "Julia": 7
}

aprovados = {nome: 'Aprovado' if nota >= 7 else 'Reprovado' for nome, nota in alunos.items()}
print(aprovados)
```

Exercício 5 - Contando palavras (Desafio)

Dada a frase:

frase = "python é incrível e python é divertido"

Utilizando Dict Comprehension, crie um dicionário onde:

a chave seja cada palavra;
o valor seja a quantidade de vezes que ela aparece.

```python
frase = "python é incrível e python é divertido"
print(frase)

contagem_palavras = {palavra: frase.count(palavra) for palavra in frase.split()}
print(contagem_palavras)
```

⭐ Desafio bônus (nível entrevista)

Dado:

funcionarios = {
    "Gabriel": 4500,
    "Lucas": 2800,
    "Ana": 5200,
    "Julia": 3100,
    "Carlos": 7000
}

Crie um novo dicionário contendo apenas os funcionários que recebem R$ 4.000 ou mais, mantendo o nome e o salário.

```python
funcionarios = {
    "Gabriel": 4500,
    "Lucas": 2800,
    "Ana": 5200,
    "Julia": 3100,
    "Carlos": 7000
}

salarios_altos = {nome: salario for nome, salario in funcionarios.items() if salario > 4000}
print(salarios_altos)
```
