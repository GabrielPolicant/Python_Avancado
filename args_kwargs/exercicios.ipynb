Exercício 1 — Somando números (*args) ⭐

```python
def soma(*args):
    total = 0
    for num in args:
        total += num
    return total

print(soma(1,4,5))
```

Exercício 2 — Encontrando o maior número (*args) ⭐⭐

```python
def maior_numero(*args):
    if not args:
        return None
    maior = -float('inf')
    for num in args:
        if num > maior:
            maior = num
    return maior

print(maior_numero(1, 67, 5, 45))
```

Exercício 3 — Cadastro de usuário (**kwargs) ⭐⭐⭐

```python
def cadastro(**kwargs):
    for item, valor in kwargs.items():
        print(f"{item.title()}: {valor.title()}")

cadastro(
    nome='Gabriel',
    sobrenome='Policant',
    idade='22',
    cidade='Assis'
)
```

Exercício 4 — Nota fiscal (*args + **kwargs) ⭐⭐⭐⭐

```python
def nota_fiscal(cliente, *args, **kwargs):
    print(f'Cliente: {cliente.title()}')
    print('')
    print('Produtos:')
    for item in args:
        print(f'- {item.title()}')
    
    print('')
    print('Informações de Pagamento:')
    for item, valor in kwargs.items():
        print(f'{item.title()} -> {valor.title()}')

nota_fiscal(
    'Gabriel',
    'Notebook',
    'Mouse',
    'Teclado',
    pagamento='PIX',
    desconto='150'
)
```

Exercício 5 — Sistema de pedidos (desafio) ⭐⭐⭐⭐⭐

```python
def pedido(numero_pedido, *args, **kwargs):
    print(f"Número do Pedido: {numero_pedido}.")
    print('')

    print('Produtos:')
    cont = 0
    for item in args:
        cont += 1
        print(f'{cont} - {item.title()}')
    
    print('')
    print(f'Quantidade de produtos: {cont}')

    print('')
    print('Dados Adicionais:')
    for item, valor in kwargs.items():
        print(f'{item.title()}: {valor.title()}')
    
pedido(
    12345,
    'Notebook',
    'Mouse',
    'Teclado',
    pagamento='PIX',
    desconto='150'
)
```

Exercício 6 — Somando valores com *args

Crie uma função chamada media() que receba qualquer quantidade de números utilizando *args e retorne a média deles.

```python
def media(*args):
    if not args:
        return 0
    total = sum(args)
    tam = len(args)
    return total / tam

print(media(1,2))
```

Crie uma função chamada menor_numero() utilizando *args.

Ela deve retornar o menor valor recebido.

```python
def menor(*args):
    if not args:
        return None

    menor = float('inf')
    for arg in args:
        if arg < menor:
            menor = arg
    return menor

print(menor(5, 2, 3, 4, 5))
```

Exercício 9 — Calculando preço final

Crie uma função chamada preco_final().

Ela deve receber:

um preço inicial;
vários descontos utilizando *args.

Cada desconto representa um valor em reais.

```python
def preco_final(valor, *args):
    total = valor
    for arg in args:
        total -= arg
    return total

print(preco_final(1000, 150, 50, 100))
```

Exercício 10 — Misturando parâmetros, *args e **kwargs

Crie uma função chamada pedido() com a seguinte assinatura:

```python
def pedido(cliente, *pedido, **informacoes):
    print(f'Cliente: {cliente.title()}')
    print('')

    print('Produtos:')
    for item in pedido:
        print(f'- {item.title()}')
    print('')

    print('Informações Adicionais:')
    for item, valor in informacoes.items():
        print(f'{item.title()}: {str(valor).title()}')

pedido(
    'Gabriel',
    'Notebook',
    'Mouse',
    'Teclado',
    pagamento='PIX',
    desconto='150'
)
```

🌟 Desafio Extra

Crie uma função chamada estatisticas().

Ela deve receber qualquer quantidade de números utilizando *args e retornar um dicionário contendo:

{
    "quantidade": 5,
    "maior": 20,
    "menor": 3,
    "media": 10.4,
    "soma": 52
}

```python
def estatistica(*args):
    if not args:
        return None

    maior, menor, quantidade, media, soma = float('-inf'), float('inf'), len(args), sum(args) / len(args), sum(args)

    for arg in args:
        if arg > maior:
            maior = arg
        if arg < menor:
            menor = arg

    return {
        'maior': maior,
        'menor': menor,
        'quantidade': quantidade,
        'media': media,
        'soma': soma
    }

print(estatistica(1, 2, -3, 4, 5))
```
