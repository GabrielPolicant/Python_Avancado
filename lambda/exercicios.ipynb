Exercício 1 - Criando sua primeira Lambda

Crie uma função lambda chamada quadrado que receba um número e retorne seu quadrado.

```python
calcula_dobro = lambda num: num * 2
numero = int(input('Digite um número: '))
print(calcula_dobro(numero))
```

Exercício 2 - Calculando desconto

Você possui uma lista de preços:

precos = [100, 250, 300, 500]

Utilizando map() e lambda, crie uma nova lista contendo os preços com 15% de desconto.

```python
precos = [100, 250, 300, 500]
precos_descontos = list(map(lambda x: x * 0.85, precos))
print(precos_descontos)
```

Exercício 3 - Convertendo temperaturas

Você possui uma lista de temperaturas em Celsius.

temperaturas = [0, 10, 20, 30, 40]

Utilizando map() e lambda, converta todas para Fahrenheit.

Fórmula
F = (C × 9/5) + 32

```python
temperaturas = [0, 10, 20, 30, 40]
temp_far = list(map(lambda c: (c * 9/5) + 32, temperaturas))
print(temp_far)
```

Exercício 4 - Deixando tudo em maiúsculo

Você possui a lista:

nomes = ["gabriel", "ana", "joao", "maria"]

Utilize map() com uma função lambda para deixar todos os nomes em letras maiúsculas.

```python
nomes = ["gabriel", "ana", "joao", "maria"]
maisculos = list(map(lambda nome: nome.upper(), nomes))
print(maisculos)
```

Exercício 5 - Desafio ⭐

Até agora você utilizou map(), que serve para modificar todos os elementos de uma lista.

Agora vamos conhecer outra função muito utilizada com lambda.

Como funciona o sorted()

O sorted() serve para ordenar uma lista.

Exemplo simples:

numeros = [5, 1, 4, 2]

print(sorted(numeros))

Saída:

[1, 2, 4, 5]

Mas ele possui um parâmetro chamado key.

Esse parâmetro recebe uma função (e é aí que a lambda entra).

A função deve retornar o valor que será utilizado para fazer a ordenação.

Agora o desafio

Você possui esta lista:

frutas = [
    "banana",
    "kiwi",
    "abacaxi",
    "uva",
    "morango"
]

Utilize sorted() com lambda para ordenar as frutas do menor nome para o maior.

```python
frutas = [
    "banana",
    "kiwi",
    "abacaxi",
    "uva",
    "morango"
]

nome_frutas_ordenado = sorted(frutas, key=lambda nome: len(nome))
print(nome_frutas_ordenado)
```

🔥 Desafio Extra (sem dica)

Você possui a seguinte lista:

funcionarios = [
    ("Gabriel", 3500),
    ("Ana", 5200),
    ("Lucas", 2800),
    ("Maria", 4700)
]

Ordene essa lista pelo salário, do menor para o maior, utilizando apenas sorted() e lambda.

```python
funcionarios = [
    ("Gabriel", 3500),
    ("Ana", 5200),
    ("Lucas", 2800),
    ("Maria", 4700)
]

salarios = sorted(funcionarios, key=lambda salario: salario[1])
print(salarios)
salarios = sorted(funcionarios, key=lambda salario: salario[1], reverse=True)
print(salarios)
```
